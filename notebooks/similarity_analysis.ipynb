```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.models import Normalize

import pandas as pd
```

```python
def read_best_scores(target_experiment: str):
    return pd.read_csv(f"../logs/{target_experiment}/best_scores.csv")
```

```python
def get_similarity_to_best(df, model):
    df = df.reset_index()
    prompts = df["prompt"].tolist()
    best_prompt_idx = df["test_score"].idxmax()
    emb = model.encode(prompts)
    sim_scores = emb @ emb[best_prompt_idx].T

    df.loc[:, "sim_to_best"] = sim_scores
    return df
```

```python
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2").append(Normalize())
```

```python
df = read_best_scores("experiment_eval")
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
df_ = df[df["task"]=="agnews"]
df_ = get_similarity_to_best(df_, model)
df_[["prompt", "test_score", "sim_to_best"]]
```

```python
df_ = df[df["task"]=="sst-5"]
df_ = get_similarity_to_best(df_, model)
df_[["prompt", "test_score", "sim_to_best"]].sort_values("sim_to_best", ascending=False)
```
